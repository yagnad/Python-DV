# Regression

```python
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
df.gender[df.gender == 'male'] = 0
df.gender[df.gender == 'female'] = 1
print(df)
```

```python
import statsmodels.formula.api as sm
result = sm.ols(
formula='grade ~ age + exercise + hours',
data=df).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(
formula='grade ~ exercise + hours',
data=df).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(
formula='grade ~ gender+ age + exercise + hours',
data=df).fit()
result.summary()
```

We can see from the above result that inclusion of gender as a variable in the regression model, slightly improves the R-square of the model by 0.1%
